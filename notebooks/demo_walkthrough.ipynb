# Ergonomic Posture Demo Walkthrough

This notebook loads the demo posture outputs, compares the two classifiers, and previews generated recommendations.

```python
from pathlib import Path
import json
import pandas as pd

ROOT = Path.cwd().resolve().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
REPORTS = ROOT / 'reports' / 'demo'
MODELS = ROOT / 'models' / 'demo'

metrics = json.loads((REPORTS / 'metrics.json').read_text())
preds = pd.read_csv(REPORTS / 'posture_predictions.csv')
metrics['best_model'], MODELS
```

```python
pd.DataFrame({
    'model': ['gradient_boosting', 'knn'],
    'accuracy': [
        metrics['models']['gradient_boosting']['mean_accuracy'],
        metrics['models']['knn']['mean_accuracy'],
    ],
    'macro_f1': [
        metrics['models']['gradient_boosting']['mean_macro_f1'],
        metrics['models']['knn']['mean_macro_f1'],
    ],
}).sort_values('accuracy', ascending=False)
```
